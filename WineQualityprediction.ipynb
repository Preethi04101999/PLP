### **Objective**
The objective of wine quality prediction is to accurately forecast wine quality based on its physicochemical properties, aiding winemakers in optimizing production processes and ensuring consistency. This also facilitates automated quality control, reducing reliance on subjective human evaluations. Ultimately, it enhances consumer satisfaction and guides informed purchase decisions.

### Import ***Library***

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/WhiteWineQuality.csv',sep=';')
```

## **Define target variable (y) and feature variables(X)**

```python
#Get Unique values (class or label in Y variable
df['quality'].value_counts()
```

```python
df.groupby('quality').mean()
```

```python
y=df['quality']
```

```python
y.shape
```

```python
y
```

```python
X=df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
```

```python
#or use .drop function to define x
X=df.drop(['quality'],axis=1)
```

```python
X
```

### **Get X variables standardized**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python

X=ss.fit_transform(X)
```

```python

X
```

### **Get Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify =y, random_state=2529)
X_train.shape,X_test.shape, y_train.shape, y_test.shape
```

### **Get Model Train**

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
```

### **Get Model Prediction**

```python

y_pred = svc.predict(X_test)
y_pred.shape
```

```python
y_pred
```

### **Get model Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
